```python
tab1[tab1.p_value < 0.05]
```

```python
vn = tab1.loc[tab1.p_value < 0.05,'Variables']
vn = vn.values
#vn = np.append(['noshowflag'],[vn])
vn
```

```python
df[vn].head()
```

```python
df2 = df.copy()
df2 = df2.dropna()
X = df2[vn]
y = df2['noshowflag']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

glm = LogisticRegression()

glm.fit(X=X,y=y)

```

```python
pred = glm.predict(X)

```

```python
score = glm.score(X, y)
print(score)
```

```python
#print(auc(X,y))
print(roc_auc_score(y.astype('int64'),pred))
```
